```python
import seaborn as sns
import matplotlib.pyplot as plt


dosya_yolu = 'AB-VERI-DUZENLENMISS.csv'

# CSV 
df = pd.read_csv(dosya_yolu)

# '25_ Fırsat_Zorluk' sütununun frekans tablosunu oluşturma
frekans_tablosu = df['25_ Fırsat_Zorluk'].value_counts().reset_index()
frekans_tablosu.columns = ['Fırsat/Zorluk', 'Frekans']


plt.figure(figsize=(10, 6))
sns.barplot(x='Fırsat/Zorluk', y='Frekans', data=frekans_tablosu, palette='Blues_d')


plt.title('25_ Fırsat_Zorluk Frekans Dağılımı', fontsize=14)
plt.xlabel('Fırsat/Zorluk', fontsize=12)
plt.ylabel('Frekans', fontsize=12)

# X eksenindeki etiketlerin düzenlenmesi
plt.xticks(rotation=45, ha='right')

# Grafik gösterimi
plt.tight_layout()  # Grafik alanının verimli kullanılmasını sağlar
plt.show()
```

```python
# Frekans
print(frekans_tablosu.to_string(index=False))
```

```python
# 'Firma Statüsü' ve '25_ Fırsat_Zorluk' sütunlarına göre gruplama yapma
frekans_gruplu = df.groupby(['Firma Statüsü', '25_ Fırsat_Zorluk']).size().unstack(fill_value=0)

# Frekans tablosunu yazdırma
print(frekans_gruplu)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Isı haritası (heatmap) ile frekansları görselleştirme
plt.figure(figsize=(14, 10))  # Grafik boyutunu büyütüyoruz, daha geniş bir alan

# Seaborn heatmap, annot=True: hücrelerin içini dolduruyor, fmt='d': tam sayılar
sns.heatmap(frekans_gruplu, annot=True, cmap='Blues', fmt='d', linewidths=0.8, linecolor='white', cbar_kws={'label': 'Frekans'}, 
            annot_kws={"size": 12, 'weight': 'bold', 'color': 'black'})

# Başlık ve etiketler
plt.title('Firma Statüsüne Göre 25_ Fırsat_Zorluk Frekans Dağılımı', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Fırsat/Zorluk Kategorisi', fontsize=14)
plt.ylabel('Firma Statüsü', fontsize=14)

# Eksen etiketlerinin düzenlenmesi
plt.xticks(fontsize=12, rotation=45, ha='right')  # X ekseni etiketleri 45 derece döndürülüp büyütülür
plt.yticks(fontsize=12)  # Y ekseni etiketlerini büyütüyoruz

# Grafik yerleşimini optimize etme
plt.tight_layout()  # Grafik elemanlarının doğru yerleşmesini
```

```python
# Sütun isimlerini kontrol edelim
print(df.columns)

# Bu çıktıyı alarak doğru sütun adını bulup kullanabilirsiniz.
```

```python
# Kategoriler için anahtar kelimeler
etiketler = {
    'Etki': ['verim kaybı', 'düşüş', 'azalma', 'tehlike', 'olumsuz', 'risk'],
    'Fırsat': ['yeni fırsat', 'artış', 'gelişme', 'potansiyel', 'güçlendirme'],
    'Zorluk': ['zorluk', 'engeller', 'sertifikasyon', 'maliyet', 'süreç', 'karşılaşmak'],
    'Öneri': ['öneri', 'planlama', 'düzenleme', 'strateji', 'çözüm']
}

# Kategorilere ayırma fonksiyonu
def kategori_ayir(cevap):
    kategoriler = {'Etki': 0, 'Fırsat': 0, 'Zorluk': 0, 'Öneri': 0}
    
    for kategori, anahtar_kelime_listesi in etiketler.items():
        for kelime in anahtar_kelime_listesi:
            if kelime.lower() in str(cevap).lower():
                kategoriler[kategori] += 1
    
    # En fazla bulunan kategori
    max_kategori = max(kategoriler, key=kategoriler.get)
    return max_kategori

# Her sütun için aynı işlemi yaparak kategori ekleyelim
for sutun in df.columns:
    if df[sutun].dtype == 'object':  # Sadece metin içeren sütunlar
        df[f'{sutun}_Kategori'] = df[sutun].apply(lambda x: kategori_ayir(x))

# Sonuçları yazdırma
print(df.head())
```

```python
# Sayısal verileri seçelim
df_numeric = df.select_dtypes(include=['float64', 'int64'])

# Korelasyon hesaplama
correlation_matrix = df_numeric.corr()

# Korelasyon ısı haritası
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Sayısal Veriler Arası Korelasyon Isı Haritası', fontsize=14)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Firma Statüsü sütunundaki değerleri kategorilere ayırma
df['Firma Statüsü'] = df['Firma Statüsü'].replace({
    'Üretici / İhracatçı, İhracatçı': 'Üretici / İhracatçı',
    'Üretici / İhracatçı': 'Üretici / İhracatçı',
    'İhracatçı': 'İhracatçı',
    'Diğer': 'Diğer'
})

# Grafik boyutunu artırma ve countplot oluşturma
plt.figure(figsize=(14, 10))  # Daha büyük grafik boyutu
sns.countplot(x='Firma Statüsü', data=df, palette='Blues', edgecolor='black')

# Başlık ve etiketler
plt.title('Firma Statüsü Dağılımı', fontsize=18, fontweight='bold')
plt.xlabel('Firma Statüsü', fontsize=16)
plt.ylabel('Frekans', fontsize=16)

# Grafik gösterimi
plt.tight_layout()
plt.show()
```

```python
def kategori_ayir(cevap):
    kategoriler = {'Etki': 0, 'Fırsat': 0, 'Zorluk': 0, 'Öneri': 0}
    
    # Anahtar kelimeler
    etiketler = {
        'Etki': ['verim kaybı', 'düşüş', 'azalma', 'tehlike', 'olumsuz'],
        'Fırsat': ['yeni fırsat', 'artış', 'gelişme', 'potansiyel', 'güçlendirme'],
        'Zorluk': ['sertifikasyon', 'maliyet', 'engel', 'üretim zorluğu', 'süreç'],
        'Öneri': ['planlama', 'strateji', 'çözüm', 'iyileştirme']
    }
    
    # Anahtar kelimelere göre kategoriler
    for kategori, anahtar_kelime_listesi in etiketler.items():
        for kelime in anahtar_kelime_listesi:
            if kelime.lower() in str(cevap).lower():
                kategoriler[kategori] += 1
    
    # En fazla bulunan kategori
    max_kategori = max(kategoriler, key=kategoriler.get)
    return max_kategori

# Veriyi kategorilere ayırma
df['Kategori'] = df['25_ Fırsat_Zorluk'].apply(kategori_ayir)

# Sonuçları yazdırma
print(df[['25_ Fırsat_Zorluk', 'Kategori']].head())
```
